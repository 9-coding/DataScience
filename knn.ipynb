```python
import pandas as pd
import numpy as np

# Create Data Frame
data = {
    'height' : [158,158,158,160,160,163,163,160,163,165,165,165,168,168,168,170,170,170],
    'weight' : [ 58, 59, 63, 59, 60, 60, 61, 64, 64, 61, 62, 65, 62, 63, 66, 63, 65, 68],
    'TShirtSize' : ['M', 'M', 'M', 'M', 'M', 'M', 'M', 'L', 'L', 'L', 'L', 'L', 'L', 'L', 'L', 'L', 'L', 'L'],
}

df = pd.DataFrame(data)

# Separating feature variables from target variables
X = df[['height', 'weight']]
y = df['TShirtSize']

# Normalization
X_normalized = (X - X.mean()) / X.std()

h_normalized = (161 - X['height'].mean()) / X['height'].std()
w_normalized = (61 - X['weight'].mean()) / X['weight'].std()
new_data_normalized = np.array([[h_normalized, w_normalized]])

# distance calculates
distances = np.sqrt(np.sum((X_normalized - new_data_normalized)**2, axis=1))

def knn(k):
  print("k = ", k)
  nearest_indices = np.argsort(distances)[:k]

  # Output target variable values for neighbors
  nearest_labels = y.iloc[nearest_indices]
  # Output with index of nearest data
  # print(nearest_labels)

  # Derive the most calculated value prediction results
  prediction = nearest_labels.mode()[0] 
  print("predict : ", prediction)


# Select k nearest neighbors
knn(3)

knn(5)
```
