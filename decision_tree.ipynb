```python
import pandas as pd
import math

# Define the dataset
data = {
    'Distinct': ['Suburban', 'Suburban', 'Rural', 'Urban', 'Urban', 'Urban', 'Rural', 'Suburban', 'Suburban', 'Urban', 'Suburban', 'Rural', 'Rural', 'Urban'],
    'HouseType': ['Detached', 'Detached', 'Detached', 'Semi-detached', 'Semi-detached', 'Semi-detached', 'Semi-detached', 'Terrace', 'Semi-detached', 'Terrace', 'Terrace', 'Terrace', 'Detached', 'Terrace'],
    'Income': ['High', 'High', 'High', 'High', 'Low', 'Low', 'Low', 'High', 'Low', 'Low', 'Low', 'High', 'Low', 'High'],
    'PreviousCustomer': ['No', 'Yes', 'No', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes'],
    'Outcome': ['NotResponded', 'NotResponded', 'Responded', 'Responded', 'Responded', 'NotResponded', 'Responded', 'NotResponded', 'Responded', 'Responded', 'Responded', 'Responded', 'Responded', 'NotResponded', ]
}

column = ['Distinct', 'HouseType', 'Income', 'PreviousCustomer', 'Outcome'] 

# The entropy calculation expression of a feature with two values
def calculate2(a,b,c,d):
  total = a+b+c+d
  if (a != 0 and b != 0):
    A = -a / (a+b) * math.log2(a / (a+b)) - b / (a+b) * math.log2(b / (a+b))
  else:
    A=0
  if(c != 0 and d != 0):
    B = -c / (c+d) * math.log2(c / (c+d)) - d / (c+d) * math.log2(d / (c+d))
  else:
    B = 0
  E = (a+b) / total * A + (c+d) / total * B
  return E

# The entropy calculation expression of a feature with three values
def calculate3(a,b,c,d,e,f):
  total = a+b+c+d+e+f # 14
  if(a != 0 and b != 0):
    A = -a / (a+b) * math.log2(a / (a+b)) - b / (a+b) * math.log2(b / (a+b))
  else:
    A = 0
  if(c != 0 and d != 0):
    B = -c / (c+d) * math.log2(c / (c+d)) - d / (c+d) * math.log2(d / (c+d))
  else:
    B = 0
  if(f != 0 and e != 0):
    C = -e / (e+f) * math.log2(e / (e+f)) - f / (e+f) * math.log2(f / (e+f))
  else:
    C = 0
  E = (a+b) / total * A + (c+d) / total * B + (e+f) / total * C
  return E


# Define the function to calculate entropy of a column
def init_entropy(a, b):
  if (a != 0 and b != 0):
    E = -a / (a+b) * math.log2(a / (a+b)) - b / (a+b) * math.log2(b / (a+b))
  else:
    E = 0
  return E  

# Calculate entropy for each feature with two values.
def entropy2(column, attr1, attr2):  
  a = len(df[(df[column]==attr1) & (df['Outcome']=='Responded')])
  b = len(df[(df[column]==attr1) & (df['Outcome']=='NotResponded')])
  c = len(df[(df[column]==attr2) & (df['Outcome']=='Responded')])
  d = len(df[(df[column]==attr2) & (df['Outcome']=='NotResponded')])
  return calculate2(a,b,c,d)

# Calculate entropy for each feature with three values.
def entropy3(column, attr1, attr2, attr3):  
  a = len(df[(df[column]==attr1) & (df['Outcome']=='Responded')])
  b = len(df[(df[column]==attr1) & (df['Outcome']=='NotResponded')])
  c = len(df[(df[column]==attr2) & (df['Outcome']=='Responded')])
  d = len(df[(df[column]==attr2) & (df['Outcome']=='NotResponded')])
  e = len(df[(df[column]==attr3) & (df['Outcome']=='Responded')])
  f = len(df[(df[column]==attr3) & (df['Outcome']=='NotResponded')])
  return calculate3(a,b,c,d,e,f)

# The entropy calculation expression of a feature with two values after the root node is determined.
def entropy4(column1, attr, column2, attr1, attr2):
  a = len(df[(df[column1]==attr) & (df[column2]==attr1) & (df['Outcome']=='Responded')])
  b = len(df[(df[column1]==attr) & (df[column2]==attr1) & (df['Outcome']=='NotResponded')]) 
  c = len(df[(df[column1]==attr) & (df[column2]==attr2) & (df['Outcome']=='Responded')])
  d = len(df[(df[column1]==attr) & (df[column2]==attr2) & (df['Outcome']=='NotResponded')])
  return calculate2(a,b,c,d)

# The entropy calculation expression of a feature with three values after the root node is determined.
def entropy5(column1, attr, column2, attr1, attr2, attr3):
  a = len(df[(df[column1]==attr) & (df[column2]==attr1) & (df['Outcome']=='Responded')])
  b = len(df[(df[column1]==attr) & (df[column2]==attr1) & (df['Outcome']=='NotResponded')]) 
  c = len(df[(df[column1]==attr) & (df[column2]==attr2) & (df['Outcome']=='Responded')])
  d = len(df[(df[column1]==attr) & (df[column2]==attr2) & (df['Outcome']=='NotResponded')])
  e = len(df[(df[column1]==attr) & (df[column2]==attr3) & (df['Outcome']=='Responded')])
  f = len(df[(df[column1]==attr) & (df[column2]==attr3) & (df['Outcome']=='NotResponded')])
  return calculate3(a,b,c,d,e,f)

df = pd.DataFrame(data)
a = len(df.loc[df['Outcome'] == 'NotResponded'])
b = len(df.loc[df['Outcome'] == 'Responded'])

init_E = init_entropy(a,b)

# Find root node.
# Calculate each entropy of Distinct/HouseType/Income/PreviousCustomer
E_Distinct = entropy3('Distinct', 'Suburban', 'Rural', 'Urban')
E_HouseType = entropy3('HouseType', 'detached', 'Semi-detached', 'Terrace')
E_Income = entropy2('Income', 'High', 'Low')
E_PreviousCustomer = entropy2('PreviousCustomer', 'yes', 'No')

# Calculate each information gain of Distinct/HouseType/Income/PreviousCustomer
I_Distinct = init_E - E_Distinct
I_HouseType = init_E - E_HouseType
I_Income = init_E - E_Income
I_PreviousCustomer = init_E - E_PreviousCustomer

# Save each information gain to a list and set the largest value as the root node.
informationGain = [I_Distinct, I_HouseType, I_Income, I_PreviousCustomer]
root = column[informationGain.index(max(informationGain))]
#print(root) # result : Distinct


# Find the entropy of Distinct's data Suburban/Rural/Urban and find the required subnode
first_child = [df[root][1], df[root][2], df[root][3]]

# Calculate the entropy of each Suburban/Rural/Urban
# Suburban
a = len(df[(df[root]==df[root][1]) & (df['Outcome']=='Responded')])
b = len(df[(df[root]==df[root][1]) & (df['Outcome']=='NotResponded')])
E_Suburban = init_entropy(a,b)
if (E_Suburban == 0):
  Suburban_child = 0

# Urban
a = len(df[(df[root]==df[root][2]) & (df['Outcome']=='Responded')])
b = len(df[(df[root]==df[root][2]) & (df['Outcome']=='NotResponded')])
E_rural = init_entropy(a,b)
if(E_rural == 0):
  Rural_child = 0

# Rural
a = len(df[(df[root]==df[root][3]) & (df['Outcome']=='Responded')])
b = len(df[(df[root]==df[root][3]) & (df['Outcome']=='NotResponded')])
E_Urban = init_entropy(a,b)
if(E_Urban == 0):
  Urban_child = 0



# Calculate the information gain of each Suburban/Rural/Urban and designate the largest value as the child node.
# Suburban's information gain
I_suburbanHouseType = E_Suburban - entropy5('Distinct', 'Suburban', 'HouseType', 'Detached', 'Semi-detached', 'Terrace')
I_suburbanIncome = E_Suburban - entropy4('Distinct', 'Suburban', 'Income', 'High', 'Low')
I_suburbanPreviousCustomer = E_Suburban - entropy4('Distinct', 'Suburban', 'PreviousCustomer', 'yes', 'No')

# Save each information gain to a list and set the largest value as the child node in Suburban direction.
informationGain_Suburban = [I_suburbanHouseType, I_suburbanIncome, I_suburbanPreviousCustomer]
Suburban_child = column[informationGain_Suburban.index(max(informationGain_Suburban))+1]


# Since the entropy of the rural is zero, the calculation of the rural is finished: not included in the code.


# Urban's information gain
I_UrbanHouseType = E_Urban - entropy5('Distinct', 'Urban', 'HouseType', 'Detached', 'Semi-detached', 'Terrace')
I_UrbanIncome = E_Urban - entropy4('Distinct', 'Urban', 'Income', 'High', 'Low')
I_UrbanPreviousCustomer = E_Urban - entropy4('Distinct', 'Urban', 'PreviousCustomer', 'yes', 'No')

# Save each information gain to a list and set the largest value as the child node in Urban direction.
informationGain_Urban = [I_UrbanHouseType, I_UrbanIncome, I_UrbanPreviousCustomer]
Urban_child = column[informationGain_Urban.index(max(informationGain_Urban))+1]

# print(entropy4('Distinct', 'Suburban', 'Income', 'High', 'Low'))

a = len(df[(df[root]=='Rural') & (df['Outcome']=='Responded')])
b = len(df[(df[root]=='Rural') & (df['Outcome']=='NotResponded')])
# print(init_entropy(a,b))

# print(entropy4('Distinct', 'Urban', 'PreviousCustomer', 'Yes', 'No'))

# Since all three entropy above are zero, it can be seen that the decision tree is completed.

child = [Suburban_child, Rural_child, Urban_child]
# 0 means that no child nodes are needed in that direction.
result = [root, child]
print("Tree result")
print(result)

#Construct a decision tree based on what is found above
def decisionTree(District, HouseType, Income, PreviousCustomer):
  if (District == 'Suburban'):
    if (Income == 'High'):
      print('Not Responded')
    elif (Income == 'Low'):
      print('Responded')
    else:
      print('Please check the input value.')
  elif(District == 'Rural'):
    print('Responded')
  elif(District == 'Urban'):
    if (PreviousCustomer == 'Yes'):
      print('Not Responded')
    elif (PreviousCustomer == 'No'):
      print('Responded')
    else:
      print('Please check the input value.')
  else:
    print('Please check the input value.')

# Enter and confirm the predict value.
print('\npredict value')
print("district = suburban, house type = detached, income = low, previous customer = yes")
print('predict : ', end="")
decisionTree('Suburban', 'Detached', 'Low', 'Yes')
```
